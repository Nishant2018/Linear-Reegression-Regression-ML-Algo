```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import datasets,linear_model
import numpy as np
```

```python
Dataset = datasets.load_diabetes()
print(Dataset)
print(Dataset.keys())
```

```python
Dataset_X = Dataset.data[:,np.newaxis,2]
print(Dataset_X)
Dataset_Y = Dataset.target
print(Dataset_Y)
```

```python
Data_X_train = Dataset_X[:-30]
Data_X_test = Dataset_X[-30:]
```

```python
Data_Y_train = Dataset_Y[:-30]
Data_Y_test = Dataset_Y[-30:]
```

```python
model = linear_model.LinearRegression()
model.fit(Data_X_train,Data_Y_train)
```

```python
Data_Y_predicted = model.predict(Data_X_test)
```

```python
plt.scatter(Data_X_test,Data_Y_test)
plt.plot(Data_X_test,Data_Y_predicted)
plt.show()
```
